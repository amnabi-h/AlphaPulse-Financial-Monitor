Future Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the 'look' of our charts to be professional
plt.style.use('fivethirtyeight')
print("✅ Tools are ready!")
```

```python
# Load the math we did earlier
corr = pd.read_csv('04_correlation_matrix.csv', index_col=0)

# Make it look like a professional heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='RdYlGn', center=0)
plt.title('Asset Correlation Matrix: Finding Safety in Diversification')
plt.show()
```

```python
# Load the 10,000 futures
mc_results = pd.read_csv('monte_carlo_results.csv', index_col=0)

plt.figure(figsize=(12, 6))
# We only plot the first 100 paths so it doesn't look messy
plt.plot(mc_results.iloc[:, :100], color='blue', alpha=0.1) 
plt.axhline(y=1, color='red', linestyle='--', label='Starting Capital')
plt.title('Monte Carlo Simulation: 10,000 Possible Futures for AlphaPulse')
plt.xlabel('Days into the Future')
plt.ylabel('Portfolio Value ($)')
plt.show()
```
